## 🍵 2.2 BeautifulSoup : le parser HTML

`BeautifulSoup` lit le HTML et nous permet de **naviguer dans la structure** et **d'extraire des éléments**.

| Méthode | Description | Exemple |
|---------|-------------|--------|
| `find(balise)` | Trouve le **1er** élément | `soup.find('h2')` |
| `find_all(balise)` | Trouve **tous** les éléments | `soup.find_all('div')` |
| `find(class_='...')` | Cherche par classe CSS | `soup.find(class_='prix')` |
| `.get_text()` | Extrait le texte | `elem.get_text(strip=True)` |
| `['href']` | Lit un attribut | `a_tag['href']` |
